```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import defaultdict

# Binance deposit address
BINANCE_DEPOSIT_ADDRESS = "0xF977814e90dA44bFA03b6295A0616a897441aceC"

def load_transactions(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

def filter_binance_deposits(transactions):
    """Filter transactions where 'to' is Binance deposit address"""
    return [tx for tx in transactions 
            if tx.get('to', '').lower() == BINANCE_DEPOSIT_ADDRESS.lower()]

def analyze_token_transfers(transactions):
    # Create data structures for analysis
    token_stats = defaultdict(lambda: {
        'count': 0,
        'total_value': 0,
        'values': []
    })
    
    for tx in transactions:
        if 'contract_address' in tx:
            contract = tx['contract_address']
            amount = tx['amount']
            
            token_stats[contract]['count'] += 1
            token_stats[contract]['total_value'] += amount
            token_stats[contract]['values'].append(amount)
    
    return token_stats

def plot_token_transfers(token_stats, transactions, top_n=20):
    # Convert to DataFrame for easier plotting
    df = pd.DataFrame.from_dict(token_stats, orient='index')
    df['contract'] = df.index
    df = df.reset_index(drop=True)
    
    # Sort by total value and get top N
    df = df.sort_values('total_value', ascending=False).head(top_n)
    
    # Truncate contract addresses for better display
    df['contract_short'] = df['contract'].apply(
        lambda x: f"{x[:6]}...{x[-4:]}" if len(x) == 42 else x)
    
    # Set up plots
    plt.figure(figsize=(15, 10))
    
    # Plot 1: Total Transfer Value to Binance by Token
    plt.subplot(2, 1, 1)
    sns.barplot(x='total_value', y='contract_short', data=df, palette='viridis')
    plt.title(f'Top {top_n} Tokens Deposited to Binance by Total Value')
    plt.xlabel('Total Value Deposited')
    plt.ylabel('Token Contract')
    
    # Plot 2: Deposit Count to Binance by Token
    plt.subplot(2, 1, 2)
    sns.barplot(x='count', y='contract_short', data=df, palette='magma')
    plt.title(f'Top {top_n} Tokens Deposited to Binance by Number of Deposits')
    plt.xlabel('Number of Deposits')
    plt.ylabel('Token Contract')
    
    plt.tight_layout()
    plt.show()
    
    # Plot 3: Distribution of deposit amounts for top tokens
    if not df.empty:
        plt.figure(figsize=(15, 8))
        top_contracts = df['contract'].tolist()
        top_transfers = [tx for tx in transactions 
                        if 'contract_address' in tx and tx['contract_address'] in top_contracts]
        
        if top_transfers:
            top_df = pd.DataFrame(top_transfers)
            # Create shortened contract addresses for display
            top_df['contract_short'] = top_df['contract_address'].apply(
                lambda x: f"{x[:6]}...{x[-4:]}" if len(x) == 42 else x)
            
            sns.boxplot(x='amount', y='contract_short', data=top_df, 
                       order=df['contract_short'], palette='coolwarm')
            plt.title('Distribution of Deposit Amounts to Binance for Top Tokens')
            plt.xlabel('Deposit Amount (log scale)')
            plt.ylabel('Token Contract')
            plt.xscale('log')  # Use log scale for better visualization
            plt.tight_layout()
            plt.show()

def main():
    # Load your transaction data
    file_path = 'recent_transactions/txs_23to21_hours.json'
    transactions = load_transactions(file_path)
    
    # Filter for Binance deposits only
    binance_deposits = filter_binance_deposits(transactions)
    
    # Analyze the data
    token_stats = analyze_token_transfers(binance_deposits)
    
    # Generate plots
    plot_token_transfers(token_stats, binance_deposits, top_n=15)
    
    # Print some summary statistics
    print("\nBinance Deposit Summary:")
    print(f"Total transactions analyzed: {len(transactions)}")
    print(f"Binance deposit transactions: {len(binance_deposits)}")
    print(f"Unique token contracts deposited: {len(token_stats)}")
    
    # Show top 5 tokens by total value
    if token_stats:
        top_tokens = sorted(token_stats.items(), key=lambda x: x[1]['total_value'], reverse=True)[:5]
        print("\nTop 5 Tokens Deposited to Binance by Total Value:")
        for contract, stats in top_tokens:
            print(f"{contract}: {stats['total_value']:.2f} (in {stats['count']} deposits)")
    else:
        print("\nNo token deposits to Binance found in the data.")

if __name__ == "__main__":
    main()
```
