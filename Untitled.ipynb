```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import re
from collections import defaultdict

BINANCE_DEPOSIT_ADDRESS = "0xF977814e90dA44bFA03b6295A0616a897441aceC"

def find_transaction_files(directory="recent_transactions"):
    """Find all transaction files with date ranges"""
    pattern = r'txs_(\d{4}-\d{2}-\d{2})_to_(\d{4}-\d{2}-\d{2})_block_(\d+)_to_(\d+)\.json'
    files = []
    
    for filename in os.listdir(directory):
        match = re.match(pattern, filename)
        if match:
            start_date, end_date, start_block, end_block = match.groups()
            files.append({
                'path': os.path.join(directory, filename),
                'start_block': int(start_block),
                'end_block': int(end_block),
                'start_date': start_date,
                'end_date': end_date
            })
    
    return sorted(files, key=lambda x: x['start_block'])

def load_transactions(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)

def load_all_transactions(file_list):
    all_transactions = []
    for file_info in file_list:
        try:
            transactions = load_transactions(file_info['path'])
            all_transactions.extend(transactions)
            print(f"Loaded {len(transactions)} transactions from {file_info['start_date']} to {file_info['end_date']} (Blocks {file_info['start_block']}-{file_info['end_block']})")
        except Exception as e:
            print(f"Error loading {file_info['path']}: {e}")
    return all_transactions

def filter_binance_deposits(transactions):
    return [tx for tx in transactions 
           if tx.get('to', '').lower() == BINANCE_DEPOSIT_ADDRESS.lower()]

def analyze_token_transfers(transactions):
    token_stats = defaultdict(lambda: {
        'count': 0,
        'total_value': 0,
        'values': [],
        'first_date': None,
        'last_date': None
    })
    
    for tx in transactions:
        if 'contract_address' in tx:
            contract = tx['contract_address']
            amount = tx['amount']
            date = tx['datetime'].split()[0]
            
            token_stats[contract]['count'] += 1
            token_stats[contract]['total_value'] += amount
            token_stats[contract]['values'].append(amount)
            
            if not token_stats[contract]['first_date'] or date < token_stats[contract]['first_date']:
                token_stats[contract]['first_date'] = date
            if not token_stats[contract]['last_date'] or date > token_stats[contract]['last_date']:
                token_stats[contract]['last_date'] = date
    
    return token_stats

def plot_token_transfers(token_stats, transactions, top_n=20):
    if not token_stats:
        print("No token transfer data to plot")
        return
    
    df = pd.DataFrame.from_dict(token_stats, orient='index')
    df['contract'] = df.index
    df = df.reset_index(drop=True)
    df = df.sort_values('total_value', ascending=False).head(top_n)
    df['contract_short'] = df['contract'].apply(lambda x: f"{x[:6]}...{x[-4:]}" if len(x) == 42 else x)
    
    plt.figure(figsize=(15, 10))
    
    plt.subplot(2, 1, 1)
    sns.barplot(x='total_value', y='contract_short', data=df, palette='viridis')
    plt.title(f'Top {top_n} Tokens Deposited to Binance by Total Value')
    plt.xlabel('Total Value Deposited')
    plt.ylabel('Token Contract')
    
    plt.subplot(2, 1, 2)
    sns.barplot(x='count', y='contract_short', data=df, palette='magma')
    plt.title(f'Top {top_n} Tokens Deposited to Binance by Number of Deposits')
    plt.xlabel('Number of Deposits')
    plt.ylabel('Token Contract')
    
    plt.tight_layout()
    plt.show()
    
    if not df.empty:
        plt.figure(figsize=(15, 8))
        top_contracts = df['contract'].tolist()
        top_transfers = [tx for tx in transactions 
                        if 'contract_address' in tx and tx['contract_address'] in top_contracts]
        
        if top_transfers:
            top_df = pd.DataFrame(top_transfers)
            top_df['contract_short'] = top_df['contract_address'].apply(
                lambda x: f"{x[:6]}...{x[-4:]}" if len(x) == 42 else x)
            
            sns.boxplot(x='amount', y='contract_short', data=top_df, 
                       order=df['contract_short'], palette='coolwarm')
            plt.title('Distribution of Deposit Amounts to Binance for Top Tokens')
            plt.xlabel('Deposit Amount (log scale)')
            plt.ylabel('Token Contract')
            plt.xscale('log')
            plt.tight_layout()
            plt.show()

def main():
    try:
        print("Discovering transaction files...")
        transaction_files = find_transaction_files()
        
        if not transaction_files:
            print("No transaction files found")
            return
        
        print("\nLoading transactions...")
        all_transactions = load_all_transactions(transaction_files)
        
        if not all_transactions:
            print("No transactions loaded")
            return
        
        print("\nFiltering for Binance deposits...")
        binance_deposits = filter_binance_deposits(all_transactions)
        
        token_stats = analyze_token_transfers(binance_deposits)
        plot_token_transfers(token_stats, binance_deposits, top_n=15)
        
        print("\nBinance Deposit Summary:")
        print(f"Total transactions analyzed: {len(all_transactions)}")
        print(f"Binance deposit transactions: {len(binance_deposits)}")
        print(f"Unique token contracts deposited: {len(token_stats)}")
        
        if token_stats:
            top_tokens = sorted(token_stats.items(), key=lambda x: x[1]['total_value'], reverse=True)[:5]
            print("\nTop 5 Tokens Deposited to Binance:")
            for contract, stats in top_tokens:
                print(f"{contract}:")
                print(f"  Total Value: {stats['total_value']:.4f}")
                print(f"  Deposit Count: {stats['count']}")
                print(f"  First Seen: {stats['first_date']}")
                print(f"  Last Seen: {stats['last_date']}")
        else:
            print("\nNo token deposits to Binance found")
            
    except Exception as e:
        print(f"\n Error: {e}")

if __name__ == "__main__":
    main()
```
